```python
# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
# they may impact your analysis.






import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("flight_price.xlsx")

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Price')

# Add labels and title
plt.ylabel('Flight Price')
plt.title('Boxplot of Flight Prices')

# Show the boxplot
plt.tight_layout()
plt.show()

# Calculate potential outliers using the interquartile range (IQR)
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

potential_outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

print("Potential outliers:")
print(potential_outliers)


```

```python
# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL to the dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'

# Load the dataset
df = pd.read_csv(url)

# Convert "Price" column to numeric (handle invalid values)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Filter out free and invalid apps
valid_paid_apps = df[(df['Price'] > 0) & df['Price'].notna()]

# Group the data by app type and calculate the average price
avg_price_by_type = valid_paid_apps.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_price_by_type.index, y=avg_price_by_type.values, palette='Set2')

# Add labels and title
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type (Paid Apps Only)')

# Show the bar chart
plt.tight_layout()
plt.show()



```

```python
# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
# with the highest number of installs.



import pandas as pd

# URL to the dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'

# Load the dataset
df = pd.read_csv(url)

# Create a frequency table of app installs
install_freq = df['Installs'].value_counts()

# Convert the index to strings (e.g., '1,000,000+' to '1000000')
install_freq.index = install_freq.index.str.replace(',', '').str.replace('+', '').astype(int)

# Sort the frequency table in descending order and select the top 10
top_10_installs = install_freq.sort_index(ascending=False).head(10)

print("Top 10 most popular apps by number of installs:")
print(top_10_installs)
```
