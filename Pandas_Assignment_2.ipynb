```python
# # Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

# In pandas, the `loc` and `iloc` are used to access and retrieve data from a DataFrame, but they have some differences in their behavior:

# 1. `loc`: This function is used to access data in a DataFrame using labels or boolean indexing. It works primarily with the row and column labels. Here are some key points about `loc`:

#    - Syntax: `df.loc[row_indexer, column_indexer]`

#    - The row_indexer can be a single label, a list of labels, or a boolean array.

#    - The column_indexer can be a single label, a list of labels, or a boolean array.

#    - The indexing is inclusive, meaning both the start and end labels are included in the result.

#    - Example: `df.loc[2:4, 'column_name']` will return rows with index labels 2, 3, and 4 for the specified column.

# 2. `iloc`: This function is used to access data in a DataFrame using integer-based indexing. It works with the integer positions of the rows and columns. Here are some key points about `iloc`:

#    - Syntax: `df.iloc[row_indexer, column_indexer]`

#    - The row_indexer and column_indexer can be a single integer, a list of integers, or a boolean array.

#    - The indexing is exclusive for the end position, meaning the end position is not included in the result.

#    - Example: `df.iloc[2:4, 1]` will return the values in the second and third rows of the second column.

# In summary, `loc` uses label-based indexing, while `iloc` uses integer-based indexing. The main difference is that `loc` includes the end position in the result, while `iloc` excludes it.
```

```python
# Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
# then find the output for both new_df.loc[2] and new_df.iloc[2].

# Did you observe any difference in both the outputs? If so then explain it.
# Consider the below code to answer further questions:

# Consider the below code to answer further questions:
# import pandas as pd
# import numpy as np
# columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
# indices = [1,2,3,4,5,6]
# #Creating a dataframe:
# df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
# ========================================================================================================
                                                 # Answer: ðŸ’¦
# To reindex the given DataFrame `df1` using the variable `reindex = [3,0,1,2]`, you can use the `reindex` function in pandas. Here's the code:

import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]

# Creating a DataFrame
df1 = pd.DataFrame(np.random.rand(6,6), columns=columns, index=indices)

# Reindexing the DataFrame
reindex = [3, 0, 1, 2]
new_df = df1.reindex(reindex)

# Accessing the data using loc
print(new_df.loc[2])

# Accessing the data using iloc
print(new_df.iloc[2])




# Observations:

# - `new_df.loc[2]` returns the row with the label/index `2` in the `new_df`. In this case, since we have reindexed the DataFrame, the label `2` now corresponds to the original row with index `1` in `df1`.

# - `new_df.iloc[2]` returns the third row (index `2`) in the `new_df` based on the new index positions. In this case, the third row in `new_df` corresponds to the original row with index `1` in `df1`.

# Therefore, there is a difference in the outputs of `new_df.loc[2]` and `new_df.iloc[2]`. The former returns the row based on label-based indexing, while the latter returns the row based on integer-based indexing.
```

# Q6. What do you understand about the windows function in pandas and list the types of windows
# functions?


In pandas, the `windows` function is used to apply calculations or transformations on a sliding or rolling window of data. It allows you to perform various operations on subsets of your data, such as calculating rolling averages, cumulative sums, or applying custom functions to a specified window size.

Pandas provides several types of window functions that can be used with the `windows` function. Some of the commonly used window functions are:

1. Rolling Window Functions:
   - `rolling()`: This function creates a rolling window object that can be used to apply various calculations, such as mean, sum, min, max, etc., over a specified window size.

2. Expanding Window Functions:
   - `expanding()`: This function creates an expanding window object that expands the window size with each new data point. It is useful for calculating cumulative quantities, such as cumulative sum or cumulative mean.

3. Exponential Weighted Window Functions:
   - `ewm()`: This function provides exponential weighted functions for calculating rolling statistics. It assigns exponentially decreasing weights to the data points in the window based on their age.

4. Groupby Window Functions:
   - `groupby().rolling()`: This combination allows you to apply window functions on grouped data. It calculates window functions independently for each group.

5. Time-based Window Functions:
   - `rolling().window()`: These functions are specific to time-series data. They allow you to create a rolling window based on time intervals, such as hours, days, or months.

These are some of the types of window functions available in pandas. They provide flexible ways to analyze and manipulate data within specified windows, facilitating various data analysis tasks.

```python
# Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
# calculates the difference between them in days, hours, and minutes using Pandas time delta. The
# program should prompt the user to enter the dates and display the result.





import pandas as pd

# Prompt the user to enter the first date
date1 = input("Enter the first date (YYYY-MM-DD): ")

# Prompt the user to enter the second date
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Convert the input dates to pandas datetime objects
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate the timedelta between the two dates
timedelta = date2 - date1

# Extract the number of days, hours, and minutes from the timedelta
days = timedelta.days
hours = timedelta.seconds // 3600
minutes = (timedelta.seconds // 60) % 60

# Display the result
print("The difference between the two dates is:")
print("Days:", days)
print("Hours:", hours)
print("Minutes:", minutes)
```

```python
# Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
# column to a categorical data type. The program should prompt the user to enter the file path, column
# name, and category order, and then display the sorted data.




import pandas as pd

# Prompt the user to enter the file path
file_path = input("Enter the file path: ")

# Prompt the user to enter the column name
column_name = input("Enter the column name: ")

# Prompt the user to enter the category order
category_order = input("Enter the category order (comma-separated values): ")
category_order = category_order.split(",")

# Read the CSV file
df = pd.read_csv(file_path)

# Convert the specified column to categorical data type
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Sort the data by the specified column
df_sorted = df.sort_values(by=column_name)

# Display the sorted data
print("Sorted Data:")
print(df_sorted)

```

```python
# Q10. Write a Python program that reads a CSV file containing sales data for different products and
# visualizes the data using a stacked bar chart to show the sales of each product category over time. The
# program should prompt the user to enter the file path and display the chart.



import pandas as pd
import matplotlib.pyplot as plt

# Prompt the user to enter the file path
file_path = input("Enter the file path: ")

# Read the CSV file
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Group the data by 'Date' and 'Category', and calculate the sum of sales
grouped_data = df.groupby(['Date', 'Category'])['Sales'].sum().unstack()

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)

# Set chart labels and title
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales of Each Product Category Over Time')

# Display the chart
plt.show()
```
