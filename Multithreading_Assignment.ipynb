````python
# Q1>  what is multithreading in python? why is it used? Name the module used to handle threads in python?


# Multithreading in Python refers to the ability of a program to simultaneously execute multiple threads of execution within a single process. Threads are lightweight units of execution that allow for concurrent execution of tasks. Each thread runs independently, performing a specific set of instructions.

# Multithreading is used in Python to achieve concurrency, where multiple tasks can be executed concurrently, potentially improving the overall performance and responsiveness of a program. It allows for the execution of multiple threads within a single process, enabling tasks to run in parallel or asynchronously.

# Python provides the "threading" module as the primary module for handling threads. It is part of the standard library and offers high-level abstractions for working with threads. The "threading" module allows you to create and manage threads, synchronize access to shared resources, and coordinate thread execution.

# Here's a simple example that demonstrates the usage of multithreading in Python using the "threading" module:

# ```python
import threading

def print_numbers():
    for i in range(1, 6):
        print(i)

def print_letters():
    for letter in 'ABCDE':
        print(letter)

# Create thread objects
t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

# Start the threads
t1.start()
t2.start()

# Wait for the threads to finish
t1.join()
t2.join()

# Output:
# 1
# A
# 2
# B
# 3
# C
# 4
# D
# 5
# E
```

# In the above example, two threads, `t1` and `t2`, are created. Each thread executes a different function concurrently. The `start()` method initiates the execution of each thread, and the `join()` method ensures that the main thread waits for the completion of both threads before continuing.

# It's important to note that Python's Global Interpreter Lock (GIL) restricts true parallel execution of threads due to internal implementation details. However, multithreading can still be useful for I/O-bound tasks or when working with external processes that release the GIL. For CPU-bound tasks, multiprocessing or other concurrency models may be more appropriate.
````

````python
 #Q2> why threading module used? write the use of the following functions
 #  1.activeCount
 # 2. currentThread
 # 3.enumerate


# The threading module in Python is used to handle threads and provides a high-level interface for working with threads. It allows you to create, manage, and synchronize threads in your Python programs. Here's an explanation of the functions you mentioned:

# 1. `activeCount()`: This function is used to return the number of Thread objects currently alive. It returns the number of threads that are currently running, including the main thread. It can be helpful to track the number of active threads in your program. Here's an example:

# ```python
import threading

def my_thread_function():
    print("Hello from thread")

# Create multiple thread objects
threads = []
for _ in range(5):
    t = threading.Thread(target=my_thread_function)
    threads.append(t)
    t.start()

print("Number of active threads:", threading.activeCount())

# Output:
# Hello from thread
# Hello from thread
# Hello from thread
# Hello from thread
# Hello from thread
# Number of active threads: 6 (including the main thread)
```

# In this example, five threads are created and started. The `activeCount()` function is then used to determine the number of active threads, including the main thread.

# 2. `currentThread()`: This function is used to return the current Thread object corresponding to the calling thread. It allows you to obtain a reference to the currently executing thread. Here's an example:

# ```python
import threading

def my_thread_function():
    current_thread = threading.currentThread()
    print("Current thread name:", current_thread.name)

t = threading.Thread(target=my_thread_function)
t.start()

# Output:
# Current thread name: Thread-1 (name may vary)
```

# In this example, the `currentThread()` function is called within the `my_thread_function()`. It returns the Thread object representing the currently executing thread, which can then be used to access information such as the thread's name.

# 3. `enumerate()`: This function is used to return a list of all Thread objects currently alive. It returns a list of all active threads, including the main thread. Each thread is represented by a Thread object. Here's an example:

# ```python
import threading

def my_thread_function():
    print("Hello from thread")

# Create multiple thread objects
threads = []
for _ in range(5):
    t = threading.Thread(target=my_thread_function)
    threads.append(t)
    t.start()

# Enumerate all active threads
all_threads = threading.enumerate()
for thread in all_threads:
    print("Thread name:", thread.name)

# Output:
# Hello from thread
# Hello from thread
# Hello from thread
# Hello from thread
# Hello from thread
# Thread name: MainThread
# Thread name: Thread-1 (name may vary)
# Thread name: Thread-2 (name may vary)
# Thread name: Thread-3 (name may vary)
# Thread name: Thread-4 (name may vary)
# Thread name: Thread-5 (name may vary)
```

# In this example, five threads are created and started. The `enumerate()` function is then used to retrieve a list of all active threads, including the main thread. Each thread's name is printed using the Thread object's `name` attribute.

# These functions provide useful capabilities for managing and monitoring threads in Python programs using the threading module.
````

````python
# Q3> Explain the following functions
# 1.run()
# 2.start()
# 3. join()
# 4.isAlive()


# Sure! Here's an explanation of the functions you mentioned:

# 1. `run()`: The `run()` method is the entry point for the thread's activity. When you create a Thread object and call its `start()` method, it internally calls the `run()` method of the target function or method you provided during thread creation. You can also explicitly call the `run()` method, but it will be executed in the context of the current thread and not as a separate thread. Here's an example:

# ```python
import threading

def my_thread_function():
    print("Hello from thread")

t = threading.Thread(target=my_thread_function)

# Call run() explicitly
t.run()

# Output:
# Hello from thread
```

# In this example, the `run()` method is called explicitly on the `t` Thread object, and it executes the target function `my_thread_function()` in the current thread.

# 2. `start()`: The `start()` method is used to start a thread's activity. It initializes the thread and calls the `run()` method of the target function or method in a separate thread. The `start()` method should only be called once on a Thread object. Here's an example:

# ```python
import threading

def my_thread_function():
    print("Hello from thread")

t = threading.Thread(target=my_thread_function)

# Start the thread
t.start()

# Output:
# Hello from thread
```

In this example, the `start()` method is called on the `t` Thread object, which starts the execution of the target function `my_thread_function()` in a separate thread.

# 3. `join()`: The `join()` method is used to wait for the completion of a thread. When a thread is started using the `start()` method, the main thread continues execution while the newly started thread runs in the background. The `join()` method allows the main thread to wait for the completion of the specified thread before proceeding further. Here's an example:

# ```python
import threading

def my_thread_function():
    print("Hello from thread")

t = threading.Thread(target=my_thread_function)

# Start the thread
t.start()

# Wait for the thread to finish
t.join()

print("Thread finished")

# Output:
# Hello from thread
# Thread finished
```

# In this example, the main thread starts the `t` Thread object, and then it waits for the thread to complete using the `join()` method. Once the thread finishes execution, the main thread continues and prints "Thread finished."

# 4. `isAlive()`: The `isAlive()` method is used to check whether a thread is currently active or alive. It returns `True` if the thread is currently running or has not yet completed, and `False` otherwise. Here's an example:

# ```python
import threading
import time

def my_thread_function():
    time.sleep(2)

t = threading.Thread(target=my_thread_function)

# Start the thread
t.start()

print("Thread is alive?", t.isAlive())

# Output (after 2 seconds):
# Thread is alive? False
```

# In this example, the `isAlive()` method is called on the `t` Thread object to check if the thread is still alive. Initially, the thread is active, but after waiting for 2 seconds using `time.sleep(2)`, the thread finishes, and the `isAlive()` method returns `False`.
````

````python
# Q4> write a python program to create two threads. Thread one must print the list of squares and thread
# two must print the list of cubes 

# Certainly! Here's a Python program that creates two threads: one to print a list of squares and another to print a list of cubes.

# ```python
import threading

def print_squares():
    squares = [num ** 2 for num in range(1, 6)]
    print("List of squares:", squares)

def print_cubes():
    cubes = [num ** 3 for num in range(1, 6)]
    print("List of cubes:", cubes)

# Create thread objects
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

# Start the threads
t1.start()
t2.start()

# Wait for the threads to finish
t1.join()
t2.join()

# Output:
# List of squares: [1, 4, 9, 16, 25]
# List of cubes: [1, 8, 27, 64, 125]
```

# In this program, two threads are created: `t1` for printing the list of squares and `t2` for printing the list of cubes. The `print_squares()` function calculates and prints the squares of numbers 1 to 5, while the `print_cubes()` function calculates and prints the cubes of the same range. 

# After creating the thread objects, the `start()` method is called to initiate the execution of each thread. Then, the `join()` method is used to wait for the threads to finish before continuing with the main thread. Finally, the lists of squares and cubes are printed as output.

# Note that due to the Global Interpreter Lock (GIL) in Python, the threads won't run in true parallel. However, they will execute concurrently, resulting in interleaved output.
````
