# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.



Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot:
   - Use: Shows the relationship between two numeric variables.
   - Example: You can use a scatter plot to visualize the correlation between a student's study hours and their exam scores.

2. Line Plot:
   - Use: Displays the trend of a continuous variable over time or any other continuous dimension.
   - Example: A line plot can be used to represent the stock market performance of a specific company over several months.

3. Histogram:
   - Use: Illustrates the distribution of a single variable.
   - Example: A histogram can be employed to visualize the frequency distribution of student grades in a class.

4. Bar Plot:
   - Use: Compares categorical data across different groups or variables.
   - Example: You can use a bar plot to compare the sales performance of different products in a store.

5. Heatmap:
   - Use: Represents the correlation between multiple variables using colors in a matrix format.
   - Example: A heatmap can be used to analyze the correlation between various weather variables such as temperature, humidity, and wind speed.

These are just a few examples of the plots you can create using Seaborn. The library offers many more plot types, each useful for different types of data and analysis.

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.





import seaborn as sns

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")


```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
# Note: pclass, age, and fare are columns in the titanic dataset.



import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot for 'age' based on 'pclass'
plt.subplot(1, 2, 1)  # Create a subplot for the first box plot
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age vs. Pclass')

# Create the second box plot for 'fare' based on 'pclass'
plt.subplot(1, 2, 2)  # Create a subplot for the second box plot
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare vs. Pclass')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack')

# Set plot title and labels
plt.title('Histogram of Price with Cut')
plt.xlabel('Price')

# Display the plot
plt.show()


```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.



import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for "species" column
sns.pairplot(data=iris_data, hue='species')

# Display the plot
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.



import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix form
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set plot title and labels
plt.title("Passenger Numbers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()


```
