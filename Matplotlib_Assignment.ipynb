# Matplotlib.


# Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating static, animated, and interactive plots. It is extensively used in the scientific and data analysis communities to visualize data in a clear and informative manner.

# Here are five plots that can be created using the Pyplot module of Matplotlib:

# 1. Line Plot: A line plot displays data points connected by straight lines. It is useful for visualizing trends and patterns in data over time or across a continuous variable.

# 2. Scatter Plot: A scatter plot represents individual data points as markers on a graph. It is effective for examining the relationship between two continuous variables and identifying any patterns or correlations.

# 3. Bar Plot: A bar plot (or bar chart) presents categorical data using rectangular bars. It is suitable for comparing values across different categories or groups, making it useful for visualizing counts, frequencies, or proportions.

# 4. Histogram: A histogram is used to visualize the distribution of a continuous variable. It divides the data into bins and displays the frequency or proportion of observations falling into each bin, creating a bar-like representation.

# 5. Pie Chart: A pie chart displays data as a circular graph divided into slices, where each slice represents a category or proportion of the whole. It is helpful for illustrating the composition or distribution of a dataset in a visually appealing way.

# These are just a few examples of the plots that can be created with Matplotlib's Pyplot module. Matplotlib offers numerous other types of plots and customization options to meet diverse visualization needs.

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.





# A scatter plot is a type of plot used to visualize the relationship between two continuous variables. It represents individual data points as markers on a graph, where the position of each point is determined by its values on the x and y axes. Scatter plots are useful for identifying patterns, trends, or correlations between the variables.

# To generate and plot a scatter plot using the given code, you can use the Matplotlib library in Python. Here's an example of how you can do it:


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)

# Adding title, xlabel, and ylabel
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Displaying the plot
plt.show()

# In this code, `np.random.normal()` is used to generate random data points for `x` and `y`. The `np.random.seed(3)` sets the random seed for reproducibility. The scatter plot is created using `plt.scatter(x, y)`, where `x` and `y` are the generated data arrays. Titles, x-label, and y-label are added using `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` respectively. Finally, `plt.show()` is used to display the plot.
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])



# The `subplot()` function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple axes or plots, making it easier to display and compare multiple plots in a single figure.

# Here's an example of how you can use the `subplot()` function to draw four line plots using the given data:

import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)  # Subplot 1
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)  # Subplot 2
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)  # Subplot 3
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)  # Subplot 4
plt.plot(x4, y4)
plt.title('Line 4')

# Adjusting the layout
plt.tight_layout()

# Displaying the plots
plt.show()


# In this code, `subplot(2, 2, 1)` indicates that the figure should have a 2x2 grid of subplots, and we are currently working on the first subplot. The subsequent `subplot()` calls specify the position of each subplot within the grid.

# Each line plot is created using `plt.plot(x, y)` with the corresponding data for `x` and `y`. The `plt.title()` function is used to add titles to each subplot. The `tight_layout()` function is used to adjust the spacing between subplots for better readability. Finally, `plt.show()` is used to display the figure with all four line plots.
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])




# A bar plot, also known as a bar chart or bar graph, is a visualization tool that represents categorical data using rectangular bars. The length or height of each bar corresponds to the value or magnitude of the data being represented. Bar plots are used to compare and display the values of different categories or groups.

# Bar plots are particularly useful in the following scenarios:

# 1. Comparing Categories: Bar plots allow for easy visual comparison of the values or magnitudes of different categories. They make it simple to identify which category has the highest or lowest value.

# 2. Showing Trends: Bar plots can be used to display trends or changes in data over time or across different categories. By plotting multiple bars side by side or stacked, it becomes easy to observe patterns and identify changes.

# 3. Presenting Distributions: Bar plots can be used to visualize the distribution of data within each category or group. They are effective in presenting the frequency, proportion, or percentage of each category.

# Now, let's plot a bar plot and a horizontal bar plot using the given data:


import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")

# Horizontal bar plot
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit")
plt.ylabel("Company")

plt.show()


# In this code, the `plt.bar()` function is used to create a vertical bar plot, and the `plt.barh()` function is used to create a horizontal bar plot. The `company` array represents the categories or companies, and the `profit` array represents the corresponding profit values.

# The `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` functions are used to add a title, x-axis label, and y-axis label to the plots. The `plt.figure(figsize=(8, 6))` line is optional and allows you to set the size of the figure (width, height) in inches.

# The first `plt.bar()` call creates a vertical bar plot, and the second `plt.barh()` call creates a horizontal bar plot. Finally, `plt.show()` is used to display the plots.
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)


# A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It is used to display key statistical measures, such as the median, quartiles, and potential outliers, in a concise and easily understandable manner. Box plots are particularly useful for comparing distributions or visualizing the spread and variability of data.

# The key components of a box plot are:

# - Median: The line within the box represents the median value of the dataset, dividing it into two halves.
# - Quartiles: The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile). It contains 50% of the data.
# - Whiskers: The whiskers extend from the box and represent the range of the data, excluding any outliers. They typically extend up to a certain distance (1.5 times the IQR) from the quartiles.
# - Outliers: Individual data points that lie outside the whiskers are considered outliers and are displayed as individual points.

# Box plots are used for various purposes, including:

# 1. Identifying Skewness: Box plots can help identify the skewness or asymmetry of a distribution. If the whiskers are uneven or the median is not at the center of the box, it indicates a skewed distribution.

# 2. Comparing Distributions: Box plots are effective for comparing the distributions of different datasets or groups. By placing multiple box plots side by side, it becomes easy to observe differences in medians, spreads, or presence of outliers.

# 3. Highlighting Outliers: Box plots provide a clear visual representation of outliers, helping to identify extreme values or data points that deviate significantly from the rest of the dataset.

# Now, let's plot a box plot using the given data:


import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')

plt.show()


# In this code, the `plt.boxplot()` function is used to create the box plot. The data for the two boxes, `box1` and `box2`, are passed as a list to the function. The `labels` parameter is used to provide labels for each box. The `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` functions are used to add a title and labels to the plot.

# Finally, `plt.show()` is used to display the box plot. The box plot will show the medians, quartiles, whiskers, and any outliers present in the data.
```
