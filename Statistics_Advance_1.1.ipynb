```python
Q1. What is the Probability density function?



In probability theory and statistics, the Probability Density Function (PDF) is a fundamental concept used to describe the probability distribution of a continuous random variable. The PDF is used to specify the likelihood of the random variable falling within a particular range of values.

For a continuous random variable X, the PDF is denoted as f(x) and satisfies the following properties:

1. Non-negativity: f(x) ≥ 0 for all x in the domain of X.
2. Integration over the entire range: ∫[over the entire real line] f(x) dx = 1.

Mathematically, the probability that X lies within an interval [a, b] is given by the integral of the PDF over that interval:

P(a ≤ X ≤ b) = ∫[from a to b] f(x) dx.

The PDF describes the shape of the probability distribution of a continuous random variable. It does so by assigning higher probabilities to regions where the random variable is more likely to take on values and lower probabilities to regions where it is less likely to occur.

For example, in the case of a normal (Gaussian) distribution, the PDF is given by the well-known bell-shaped curve. Different probability distributions have different PDFs that capture their unique characteristics.

It's important to note that for discrete random variables, we use a Probability Mass Function (PMF) instead of a PDF. The PMF gives the probability of each individual value that the discrete random variable can take.
```

```python
Q2. What are the types of Probability distribution?


There are several types of probability distributions used to model different types of random variables. The choice of distribution depends on the characteristics of the data and the specific problem at hand. Some common types of probability distributions include:

1. Continuous Distributions:
   a. Normal Distribution (Gaussian Distribution): The most well-known and widely used distribution. It has a bell-shaped curve and is fully defined by its mean and standard deviation.
   b. Uniform Distribution: A constant probability over a specified interval, where all outcomes are equally likely.
   c. Exponential Distribution: Often used to model the time between events in a Poisson process, where events occur at a constant average rate.
   d. Beta Distribution: Defined on the interval [0, 1] and is commonly used to model probabilities.
   e. Gamma Distribution: Generalizes the exponential distribution and is often used in the context of waiting times.

2. Discrete Distributions:
   a. Bernoulli Distribution: Models a binary outcome, like success/failure or heads/tails of a coin toss.
   b. Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.
   c. Poisson Distribution: Describes the number of events that occur in a fixed interval of time or space, given a constant average rate of occurrence.
   d. Geometric Distribution: Represents the number of Bernoulli trials needed to achieve the first success.
   e. Negative Binomial Distribution: Generalizes the geometric distribution by counting the number of trials until a specified number of successes is reached.
   f. Hypergeometric Distribution: Describes the probability of k successes in n draws, without replacement, from a finite population of size N containing K successes.

3. Multivariate Distributions:
   a. Multinomial Distribution: An extension of the binomial distribution to more than two categories.
   b. Multivariate Normal Distribution: Generalizes the normal distribution to multiple dimensions, often used in multivariate statistics.

These are just some of the commonly encountered probability distributions. There are many other specialized distributions used for specific applications in various fields of science, engineering, and data analysis. The choice of the appropriate distribution depends on the nature of the data and the assumptions that can be made about the underlying process generating the random variables.
```

```python
# Q3. Write a Python function to calculate the probability density function of a normal distribution with
# given mean and standard deviation at a given point.



# To calculate the Probability Density Function (PDF) of a normal distribution at a given point, you can use the following Python function. We'll use the `scipy.stats` module, which provides convenient functions for working with probability distributions.

# If you don't have `scipy` installed, you can install it using `pip install scipy`.


from scipy.stats import norm

def normal_pdf(x, mean, std_dev):
#     """
#     Calculate the Probability Density Function (PDF) of a normal distribution
#     at a given point.

#     Parameters:
#         x (float): The value at which to evaluate the PDF.
#         mean (float): Mean of the normal distribution.
#         std_dev (float): Standard deviation of the normal distribution.

#     Returns:
#         float: The probability density at the given point.
#     """
    return norm.pdf(x, loc=mean, scale=std_dev)

# Example usage:
mean = 0
std_dev = 1
x_value = 1.5
pdf_value = normal_pdf(x_value, mean, std_dev)
print(f"The PDF at x={x_value} is: {pdf_value}")


# In this function, we use `norm.pdf(x, loc=mean, scale=std_dev)` from `scipy.stats` to calculate the PDF at the given point `x` for a normal distribution with the specified `mean` and `std_dev`. The `loc` parameter corresponds to the mean, and the `scale` parameter corresponds to the standard deviation of the normal distribution. The function returns the probability density at the given point `x`.
```

```python
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

The Binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has the same probability of success, denoted as "p." The distribution is characterized by two parameters: the number of trials "n" and the probability of success "p."

Properties of the Binomial distribution:

1. Discreteness: The Binomial distribution is a discrete probability distribution since it deals with a fixed number of discrete events (trials) and counts the number of successes.

2. Fixed Number of Trials: The distribution is defined for a fixed number of trials "n," where each trial is independent and has the same probability of success "p."

3. Binary Outcome: Each trial results in one of two possible outcomes: success (with probability "p") or failure (with probability "1 - p").

4. Independence: The outcomes of each trial are assumed to be independent of each other, meaning the result of one trial does not influence the outcome of subsequent trials.

5. Probability Mass Function (PMF): The probability mass function of the Binomial distribution is given by the formula: P(X = k) = C(n, k) * p^k * (1 - p)^(n - k), where X is the number of successes, k is the number of successes we want to observe, C(n, k) is the binomial coefficient (n choose k), and p^k * (1 - p)^(n - k) is the probability of obtaining k successes and (n - k) failures in any order.

Two examples of events where the Binomial distribution can be applied:

1. Coin Tossing: Consider flipping a fair coin (where p = 0.5) "n" times and counting the number of heads obtained. Each flip is an independent trial with two possible outcomes (heads or tails), and the probability of getting a head in each trial is constant. The number of heads obtained in "n" coin tosses follows a Binomial distribution.

2. Manufacturing Defects: Suppose a manufacturing process produces items, and each item has a probability "p" of being defective. If we randomly sample "n" items from the production line and count the number of defective items, the number of defective items follows a Binomial distribution. For example, if we want to know the probability of getting exactly "k" defective items in a sample of size "n," we can use the Binomial distribution to calculate it.
```

```python
# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
# and plot a histogram of the results using matplotlib.


import numpy as np
import matplotlib.pyplot as plt

def generate_binomial_sample(size, p):
    """
    Generate a random sample from a binomial distribution.

    Parameters:
        size (int): The size of the sample.
        p (float): The probability of success in each trial.

    Returns:
        numpy.ndarray: An array containing the random sample.
    """
    return np.random.binomial(n=1, p=p, size=size)

# Parameters
sample_size = 1000
probability_of_success = 0.4

# Generate random sample
sample = generate_binomial_sample(sample_size, probability_of_success)

# Plot histogram
plt.hist(sample, bins=2, rwidth=0.8, align='left', color='skyblue', edgecolor='black')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (p = {probability_of_success}, n = {sample_size})')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
# with given mean at a given point.




# To calculate the Cumulative Distribution Function (CDF) of a Poisson distribution at a given point, you can use the `scipy.stats` module, which provides convenient functions for working with probability distributions.

# If you don't have `scipy` installed, you can install it using `pip install scipy`.

# Here's the Python function to calculate the CDF of a Poisson distribution:


from scipy.stats import poisson

def poisson_cdf(x, mean):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution
    at a given point.

    Parameters:
        x (int): The value at which to evaluate the CDF.
        mean (float): Mean (λ) of the Poisson distribution.

    Returns:
        float: The cumulative probability up to the given point.
    """
    return poisson.cdf(x, mu=mean)

# Example usage:
mean = 3.5
x_value = 2
cdf_value = poisson_cdf(x_value, mean)
print(f"The CDF at x={x_value} is: {cdf_value}")


# In this function, we use `poisson.cdf(x, mu=mean)` from `scipy.stats` to calculate the CDF at the given point `x` for a Poisson distribution with the specified mean `mean` (often denoted as λ). The function returns the cumulative probability up to the given point `x`.

# The Poisson distribution is commonly used to model the number of events that occur in a fixed interval of time or space, given a constant average rate of occurrence (mean). The CDF of the Poisson distribution gives the probability that the random variable is less than or equal to the given point `x`.
```

```python
Q7. How Binomial distribution different from Poisson distribution?




The Binomial distribution and the Poisson distribution are two distinct probability distributions used to model different types of events. Here are the key differences between them:

1. Nature of Events:
   - Binomial Distribution: The Binomial distribution is used to model the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes (success or failure).
   - Poisson Distribution: The Poisson distribution is used to model the number of events that occur in a fixed interval of time or space, given a constant average rate of occurrence. It deals with the count of events in a continuous range.

2. Number of Trials/Events:
   - Binomial Distribution: The Binomial distribution requires a fixed number of trials "n," where each trial is independent and has the same probability of success "p."
   - Poisson Distribution: The Poisson distribution does not involve a fixed number of trials. It focuses on the number of events that occur in a given time or space, assuming a constant rate of occurrence.

3. Probability of Success:
   - Binomial Distribution: In the Binomial distribution, the probability of success "p" is constant across all trials.
   - Poisson Distribution: In the Poisson distribution, there is no concept of a probability of success. Instead, it is defined by the average rate of events per unit of time or space, often denoted as λ (lambda).

4. Discreteness:
   - Binomial Distribution: The Binomial distribution deals with discrete outcomes, as it counts the number of successes in a fixed number of trials.
   - Poisson Distribution: The Poisson distribution also deals with discrete outcomes since it counts the number of events that occur within a continuous range.

5. Assumptions:
   - Binomial Distribution: The Binomial distribution assumes that each trial is independent and that the probability of success is constant for all trials.
   - Poisson Distribution: The Poisson distribution assumes that events occur independently and that the rate of occurrence is constant over time or space.

6. Use Cases:
   - Binomial Distribution: The Binomial distribution is appropriate for situations where we have a fixed number of independent trials, such as coin tosses, the number of successful sales calls, or the number of defective items in a sample.
   - Poisson Distribution: The Poisson distribution is suitable for situations involving rare events or where events occur independently at a constant average rate, such as the number of customer arrivals at a service center, the number of accidents in a day, or the number of phone calls received in an hour.

In summary, the Binomial distribution models the number of successes in a fixed number of trials, while the Poisson distribution models the number of events occurring in a fixed interval of time or space with a constant rate of occurrence.
```

```python
# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
# sample mean and variance.





# To generate a random sample of size 1000 from a Poisson distribution with a mean of 5 and calculate the sample mean and variance, you can use the `numpy` library in Python. If you don't have `numpy` installed, you can do so using `pip install numpy`.

# Here's the code to achieve this:


import numpy as np

def generate_poisson_sample(size, mean):
    """
    Generate a random sample from a Poisson distribution.

    Parameters:
        size (int): The size of the sample.
        mean (float): Mean (λ) of the Poisson distribution.

    Returns:
        numpy.ndarray: An array containing the random sample.
    """
    return np.random.poisson(lam=mean, size=size)

# Parameters
sample_size = 1000
mean = 5

# Generate random sample
sample = generate_poisson_sample(sample_size, mean)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample, ddof=1)  # Use ddof=1 for sample variance (Bessel's correction)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")


# In this code, we use `numpy.random.poisson` to generate a random sample of size 1000 from a Poisson distribution with the given mean (λ=5). We then use `numpy.mean` to calculate the sample mean and `numpy.var` with `ddof=1` to calculate the sample variance. The `ddof=1` argument is used for Bessel's correction to compute the unbiased sample variance. The sample mean and variance will give us an estimate of the population mean and variance based on the random sample.
```

```python
In both the Binomial distribution and the Poisson distribution, the mean and variance are closely related, but the relationship is slightly different for each distribution.

Binomial Distribution:
For a Binomial distribution with parameters "n" (number of trials) and "p" (probability of success in each trial), the mean (μ) and variance (σ^2) are related as follows:

Mean (μ) = n * p
Variance (σ^2) = n * p * (1 - p)

In the Binomial distribution, the variance is directly dependent on the number of trials (n) and the probability of success (p). The larger the number of trials or the probability of success, the larger the variance will be. The variance is maximized when p = 0.5 (in this case, the distribution is symmetric).

Poisson Distribution:
For a Poisson distribution with parameter "λ" (lambda) representing the average rate of events occurring in a fixed interval of time or space, the mean (μ) and variance (σ^2) are related as follows:

Mean (μ) = λ
Variance (σ^2) = λ

In the Poisson distribution, the mean (λ) is equal to the variance (σ^2). This unique property makes the Poisson distribution special. The variance of a Poisson distribution is solely determined by the mean rate of occurrence (λ), and it does not depend on any other parameters.

Summary:
In summary, the mean and variance of both the Binomial and Poisson distributions are related to the parameters that define each distribution. In the Binomial distribution, the variance depends on both the number of trials (n) and the probability of success (p). In contrast, the Poisson distribution has a constant variance equal to its mean rate of occurrence (λ), which is a distinctive feature of this distribution.
```
