```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.



Estimation statistics is a branch of statistics that involves making inferences and predictions about population parameters based on sample data. The goal of estimation is to estimate unknown population parameters, such as the population mean or proportion, using information obtained from a sample.

There are two main types of estimation in statistics: point estimation and interval estimation.

1. Point Estimate:
A point estimate is a single value that is used to approximate an unknown population parameter. It is computed from sample data and serves as the best guess or "point" estimate of the population parameter. Point estimates are commonly used when we want a quick and straightforward estimate of the population parameter.

For example:
- If we want to estimate the average height of all students in a school, we can randomly select a sample of students, measure their heights, and calculate the sample mean height. The sample mean height would be the point estimate of the population mean height.

Point estimates are simple to calculate and easy to interpret, but they do not provide information about the accuracy or uncertainty of the estimate. Interval estimation addresses this limitation.

2. Interval Estimate:
Interval estimation provides a range of values within which the unknown population parameter is likely to lie, along with a certain level of confidence. Instead of providing a single point estimate, interval estimates give a range of plausible values.

The interval estimate is often expressed in the form of a confidence interval. A confidence interval provides an interval estimate with a specified level of confidence, such as 95% confidence. This means that if we were to repeat the sampling process multiple times, about 95% of the resulting confidence intervals would contain the true population parameter.

For example:
- If we calculate a 95% confidence interval for the population mean height of students in a school, the interval might be [160 cm, 170 cm]. This would mean that we are 95% confident that the true population mean height lies between 160 cm and 170 cm.

Interval estimates are more informative than point estimates because they account for the uncertainty associated with the estimate. They provide a range of values that captures the likely variability of the population parameter based on the sample data.

In summary, estimation statistics involves making educated guesses about unknown population parameters based on sample data. Point estimates provide single value approximations, while interval estimates provide a range of plausible values along with a confidence level, which gives a better understanding of the accuracy and uncertainty of the estimate.
```

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard
# deviation.


def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    """
    Function to estimate the population mean using a sample mean and standard deviation.

    Parameters:
        sample_mean (float): The mean of the sample.
        sample_std_dev (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.

    Returns:
        float: Estimated population mean.
    """
    import math

    # Calculate the standard error (standard deviation of the sample mean)
    standard_error = sample_std_dev / math.sqrt(sample_size)

    # Calculate the margin of error (1.96 corresponds to a 95% confidence interval for large sample sizes)
    margin_of_error = 1.96 * standard_error

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # The estimated population mean lies within the confidence interval
    estimated_population_mean = (lower_bound + upper_bound) / 2

    return estimated_population_mean

# Example usage:
sample_mean = 75.5
sample_std_dev = 8.2
sample_size = 100
estimated_mean = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print("Estimated population mean:", estimated_mean)
```

```python
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.


Hypothesis testing is a statistical method used to make decisions and draw conclusions about a population based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (H1), and then using sample data to determine which hypothesis is more likely to be true.

The process of hypothesis testing involves the following steps:

1. Formulating Hypotheses:
   - Null Hypothesis (H0): It represents the default assumption or the status quo. It states that there is no significant difference or effect in the population, or that any observed difference is due to chance.
   - Alternative Hypothesis (H1): It represents the claim or the research hypothesis. It states that there is a significant difference or effect in the population, which is not due to chance.

2. Selecting a Test Statistic:
   - A test statistic is a quantity derived from the sample data that will be used to assess the strength of evidence against the null hypothesis.

3. Setting the Significance Level (α):
   - The significance level, denoted by α (alpha), is the probability of rejecting the null hypothesis when it is true. Commonly used significance levels are 0.05 (5%) or 0.01 (1%).

4. Computing the P-value:
   - The P-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. It measures the strength of evidence against the null hypothesis.

5. Making a Decision:
   - Based on the P-value and the significance level, we either reject the null hypothesis in favor of the alternative hypothesis if the P-value is less than or equal to α, or fail to reject the null hypothesis if the P-value is greater than α.

Importance of Hypothesis Testing:

1. Scientific Inference: Hypothesis testing provides a systematic and objective approach to testing scientific hypotheses and making data-driven decisions.

2. Validation of Claims: Hypothesis testing allows researchers to assess whether there is sufficient evidence to support a claim or a research hypothesis based on sample data.

3. Decision-Making: Hypothesis testing aids in decision-making by providing a framework to determine the best course of action, such as choosing between two competing strategies or treatments.

4. Generalizability: Hypothesis testing helps in generalizing the results from a sample to the larger population, providing insights beyond the sample data.

5. Objectivity: Hypothesis testing provides an objective and quantitative approach to evaluate the evidence against the null hypothesis, reducing biases and subjectivity in decision-making.

6. Statistical Inference: Hypothesis testing is a fundamental tool for statistical inference, which involves drawing conclusions about a population based on sample data.

In summary, hypothesis testing is a powerful statistical method used to make informed decisions and inferences based on sample data. It is widely used in scientific research, business analysis, and various fields to validate claims, assess the strength of evidence, and draw conclusions about populations.
```

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.




# To conduct a hypothesis test on the difference between two population means, we can use a t-test for independent samples (two-sample t-test) if the samples are independent or a paired t-test if the samples are dependent (e.g., paired measurements before and after an intervention).

# Here's a Python script to conduct a two-sample t-test for independent samples:


import numpy as np
from scipy.stats import ttest_ind

def hypothesis_test_two_means(sample1, sample2, alpha=0.05):
    """
    Function to conduct a hypothesis test on the difference between two population means.

    Parameters:
        sample1 (array-like): The first sample.
        sample2 (array-like): The second sample.
        alpha (float): The significance level (default is 0.05).

    Returns:
        tuple: t-statistic, p-value, and whether the null hypothesis is rejected.
    """
    # Conduct the t-test for independent samples
    t_statistic, p_value = ttest_ind(sample1, sample2)

    # Determine whether the null hypothesis is rejected based on the significance level
    reject_null_hypothesis = p_value < alpha

    return t_statistic, p_value, reject_null_hypothesis

# Example usage:
sample1 = np.array([65, 68, 70, 72, 67, 69, 71, 73, 66, 68])
sample2 = np.array([60, 62, 64, 63, 61, 62, 65, 63, 62, 64])

t_stat, p_val, reject_null = hypothesis_test_two_means(sample1, sample2)

print("t-statistic:", t_stat)
print("p-value:", p_val)
print("Reject Null Hypothesis:", reject_null)


# In this script, we use the `ttest_ind` function from the `scipy.stats` module to conduct the two-sample t-test for independent samples. The function takes two samples as input and returns the t-statistic and the p-value.

# The significance level (alpha) is set to 0.05 by default, but you can change it to any other desired value.

# After conducting the t-test, the script determines whether the null hypothesis is rejected based on the significance level. If the p-value is less than the significance level (alpha), the null hypothesis is rejected, indicating that there is evidence of a significant difference between the two population means.

# Please note that the above script assumes that the two samples are independent. If the samples are dependent, you can use the `ttest_rel` function from the `scipy.stats` module to conduct a paired t-test. The usage of `ttest_rel` is similar to `ttest_ind`, but it takes paired samples as input.
```

```python
Q6: What is a null and alternative hypothesis? Give some examples.








In hypothesis testing, the null hypothesis (H0) and alternative hypothesis (H1) are two competing statements used to make inferences about a population parameter based on sample data. These hypotheses are formulated before conducting a statistical test, and the goal is to determine which hypothesis provides better support for the observed data.

1. Null Hypothesis (H0):
The null hypothesis represents the default assumption or status quo. It states that there is no significant difference, effect, or relationship in the population. In other words, any observed difference or effect in the sample data is due to random chance or sampling variability. The null hypothesis is typically denoted by "H0."

Examples of Null Hypotheses:
- The average height of male college students is equal to the average height of female college students.
- There is no difference in the test scores of two groups of students who received different teaching methods.
- The new drug has no effect on reducing blood pressure compared to the placebo.

2. Alternative Hypothesis (H1):
The alternative hypothesis represents the claim or research hypothesis. It states that there is a significant difference, effect, or relationship in the population that is not due to random chance. The alternative hypothesis contradicts the null hypothesis and suggests that the observed data is not a result of random variability alone. The alternative hypothesis is typically denoted by "H1."

Examples of Alternative Hypotheses:
- The average height of male college students is greater than the average height of female college students.
- The test scores of students who received a new teaching method are higher than the test scores of students who received the standard teaching method.
- The new drug has a significant effect on reducing blood pressure compared to the placebo.

In hypothesis testing, the goal is to determine whether the data provides sufficient evidence to reject the null hypothesis in favor of the alternative hypothesis. This is done by conducting a statistical test (e.g., t-test, z-test, chi-square test) and calculating a p-value, which measures the strength of evidence against the null hypothesis. If the p-value is smaller than a chosen significance level (often 0.05), the null hypothesis is rejected, and the alternative hypothesis is supported, indicating that there is significant evidence to support the claim or effect stated in the alternative hypothesis. If the p-value is greater than the significance level, the null hypothesis is not rejected, and there is not enough evidence to support the claim in the alternative hypothesis.
```

```python
Q7: Write down the steps involved in hypothesis testing.




Hypothesis testing involves a series of steps to make decisions and draw conclusions about population parameters based on sample data. The steps involved in hypothesis testing are as follows:

1. Formulate the Hypotheses:
   - Null Hypothesis (H0): State the default assumption or the status quo. It represents the belief that there is no significant difference, effect, or relationship in the population.
   - Alternative Hypothesis (H1): State the research hypothesis or the claim. It represents the belief that there is a significant difference, effect, or relationship in the population that contradicts the null hypothesis.

2. Choose the Test Statistic:
   - Select an appropriate test statistic based on the nature of the data and the type of hypotheses being tested. Common test statistics include t-statistic, z-statistic, chi-square statistic, etc.

3. Set the Significance Level (α):
   - Determine the desired level of significance (alpha) to make a decision. Commonly used significance levels are 0.05 (5%) or 0.01 (1%). The significance level represents the probability of rejecting the null hypothesis when it is true.

4. Collect Sample Data:
   - Obtain a representative sample from the population of interest. The sample should be randomly selected and adequately sized to provide reliable results.

5. Compute the Test Statistic:
   - Use the sample data to calculate the value of the chosen test statistic. The test statistic measures the difference between the sample data and the expected values under the null hypothesis.

6. Calculate the P-value:
   - Determine the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. The p-value measures the strength of evidence against the null hypothesis.

7. Compare P-value with α:
   - Compare the calculated p-value with the chosen significance level (α). If the p-value is less than or equal to α, it suggests strong evidence against the null hypothesis, and we reject the null hypothesis in favor of the alternative hypothesis. If the p-value is greater than α, we fail to reject the null hypothesis.

8. Draw a Conclusion:
   - Based on the comparison of the p-value with α, make a decision and draw a conclusion about the hypotheses. If the null hypothesis is rejected, support the claim in the alternative hypothesis. If the null hypothesis is not rejected, there is insufficient evidence to support the claim in the alternative hypothesis.

9. Interpretation:
   - Interpret the results in the context of the problem and draw meaningful conclusions about the population parameter based on the sample data.

It is crucial to follow these steps systematically to ensure valid and reliable hypothesis testing results. Hypothesis testing helps in making data-driven decisions, validating claims, and drawing inferences about population parameters based on sample data.
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.



The p-value, short for "probability value," is a crucial concept in hypothesis testing. It is a probability that measures the strength of evidence against the null hypothesis (H0) based on the observed sample data. The p-value quantifies the likelihood of observing a test statistic as extreme as, or more extreme than, the one obtained from the sample, assuming the null hypothesis is true.

In hypothesis testing, the p-value serves as the basis for making decisions about the null hypothesis. Here's how the p-value is used in hypothesis testing:

1. Formulate the Hypotheses:
   - Null Hypothesis (H0): Assume that there is no significant difference, effect, or relationship in the population.
   - Alternative Hypothesis (H1): Assume that there is a significant difference, effect, or relationship in the population that contradicts the null hypothesis.

2. Calculate the Test Statistic:
   - Choose an appropriate test statistic based on the nature of the data and the type of hypotheses being tested. Calculate the test statistic using the sample data.

3. Calculate the P-value:
   - Using the test statistic and the assumed null hypothesis, calculate the p-value. The p-value represents the probability of observing the data or more extreme data points, assuming the null hypothesis is true.

4. Compare P-value with Significance Level (α):
   - Compare the calculated p-value with a pre-defined significance level (α), commonly set at 0.05 (5%) or 0.01 (1%). The significance level represents the threshold for rejecting the null hypothesis.

5. Draw a Conclusion:
   - If the p-value is less than or equal to the significance level (α), it indicates that the observed data is unlikely to occur under the assumption of the null hypothesis. In such a case, the null hypothesis is rejected in favor of the alternative hypothesis. We conclude that there is significant evidence to support the claim in the alternative hypothesis.
   - If the p-value is greater than the significance level (α), it suggests that the observed data is reasonably likely to occur by chance, even if the null hypothesis is true. In this scenario, we fail to reject the null hypothesis, and there is insufficient evidence to support the claim in the alternative hypothesis.

Significance of the P-value in Hypothesis Testing:
- The p-value allows researchers to objectively quantify the strength of evidence against the null hypothesis. It provides a continuous measure of evidence, allowing for more nuanced decision-making.
- A smaller p-value indicates stronger evidence against the null hypothesis, supporting the claim in the alternative hypothesis.
- A larger p-value suggests weaker evidence against the null hypothesis, indicating that the observed data is reasonably consistent with the null hypothesis.
- The p-value helps avoid subjectivity in hypothesis testing. Researchers can use the p-value to make objective decisions based on statistical evidence rather than personal opinions or biases.
- By comparing the p-value to a pre-defined significance level, hypothesis testing provides a clear criterion for decision-making, facilitating the communication of results and conclusions.

In summary, the p-value plays a central role in hypothesis testing by quantifying the strength of evidence against the null hypothesis. It allows researchers to make objective decisions about the null hypothesis based on the observed sample data, leading to meaningful inferences and conclusions.

```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.


import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2):
    """
    Function to calculate the two-sample t-test for independent samples.

    Parameters:
        sample1 (array-like): The first sample.
        sample2 (array-like): The second sample.

    Returns:
        float: t-statistic
        float: p-value
    """
    # Perform the two-sample t-test
    t_statistic, p_value = ttest_ind(sample1, sample2)

    return t_statistic, p_value

# Example usage:
sample1 = np.array([22, 25, 28, 30, 26])
sample2 = np.array([20, 24, 23, 26, 27])

t_stat, p_val = two_sample_t_test(sample1, sample2)

print("Two-Sample T-Test:")
print("t-statistic:", t_stat)
print("p-value:", p_val)

```

```python
Q11: What is Student’s t distribution? When to use the t-Distribution.


Student's t-distribution, often simply referred to as the t-distribution, is a probability distribution that arises in statistics and is used for making inferences about the population mean when the sample size is small or when the population standard deviation is unknown. It is a variation of the standard normal distribution but with heavier tails, allowing for more uncertainty in small samples.

The t-distribution is characterized by a parameter called degrees of freedom (df), which determines its shape. As the degrees of freedom increase, the t-distribution approaches the shape of the standard normal distribution.

The t-distribution is used in situations where the following conditions are met:

1. Small Sample Size: When dealing with a small sample size (typically n < 30) and the population standard deviation is unknown. In such cases, the t-distribution provides a better approximation than the standard normal distribution for making inferences about the population mean.

2. Population Standard Deviation Unknown: When the population standard deviation is unknown, and it needs to be estimated from the sample data. In practice, the sample standard deviation is used as an estimate for the population standard deviation, which leads to the use of the t-distribution.

Common scenarios where the t-distribution is used include:

- When conducting hypothesis tests or calculating confidence intervals for the population mean based on a small sample size.
- In situations where the standard normal distribution assumption does not hold due to limited sample data, and the t-distribution provides a more accurate approximation.

It's important to note that as the sample size increases (n ≥ 30), the t-distribution approaches the standard normal distribution. Therefore, for larger sample sizes, the t-distribution and the standard normal distribution yield similar results, and the standard normal distribution can be used instead.

In summary, the t-distribution is a probability distribution used in statistics for making inferences about the population mean when the sample size is small or when the population standard deviation is unknown. It is particularly useful when dealing with small sample sizes and situations where the standard normal distribution assumption is not appropriate. As the sample size increases, the t-distribution approaches the standard normal distribution, and the two distributions yield similar results.
```

```python
Q12: What is t-statistic? State the formula for t-statistic.






The t-statistic is a measure used in hypothesis testing to assess the difference between the sample mean and the hypothesized population mean, taking into account the variability in the sample data. It quantifies how many standard errors the sample mean is away from the hypothesized population mean. The t-statistic is a critical component in conducting t-tests and determining the significance of observed differences or effects in sample data.

The formula for calculating the t-statistic for a sample mean is as follows:

t = (x̄ - μ) / (s / √n)

Where:
- t is the t-statistic.
- x̄ is the sample mean.
- μ is the hypothesized population mean (null hypothesis).
- s is the sample standard deviation.
- n is the sample size.

In the formula, (x̄ - μ) represents the difference between the sample mean and the hypothesized population mean. The denominator (s / √n) is the standard error of the sample mean, which accounts for the variability in the sample data and provides a measure of how much the sample mean might vary from one sample to another.

The t-statistic is essentially the z-score of the sample mean, adjusted for the sample size and variability. It follows a t-distribution with (n - 1) degrees of freedom when the sample is drawn from a normally distributed population.

When conducting a hypothesis test, the t-statistic is used to calculate the p-value, which measures the probability of observing the data (or more extreme) if the null hypothesis is true. The p-value is then compared to a significance level (commonly set at 0.05) to make a decision about whether to reject or fail to reject the null hypothesis.

In summary, the t-statistic is a measure that indicates how many standard errors the sample mean is away from the hypothesized population mean. It is used in hypothesis testing to assess the significance of observed differences in sample data and plays a central role in t-tests and other statistical analyses involving small sample sizes and unknown population standard deviations.
```

```python
# Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
# sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
# Estimate the population mean revenue with a 95% confidence interval.


# To estimate the population mean revenue with a 95% confidence interval, we can use the formula for the confidence interval for the population mean based on the t-distribution. Since the sample size is relatively large (n = 50), we can use the t-distribution with (n-1) degrees of freedom to account for the uncertainty in estimating the population mean.

# The formula for the confidence interval for the population mean (μ) is:

# Confidence Interval = sample_mean ± t_critical * (sample_std_dev / √n)

# Where:
# - sample_mean = $500 (sample mean revenue)
# - sample_std_dev = $50 (standard deviation of the sample revenue)
# - n = 50 (sample size)
# - t_critical = t-score corresponding to the desired confidence level (95% confidence level in this case) with (n-1) degrees of freedom.

# To find the t_critical value for a 95% confidence level with (n-1) degrees of freedom, you can consult the t-distribution table or use a statistical tool. For a sample size of 50 and a 95% confidence level, the t_critical value is approximately 2.0096.

# Now, let's calculate the confidence interval:


import numpy as np
from scipy.stats import t

# Given data
sample_mean = 500
sample_std_dev = 50
sample_size = 50

# Calculate t_critical value for 95% confidence level with (n-1) degrees of freedom
t_critical = t.ppf(0.975, df=sample_size-1)

# Calculate the margin of error
margin_of_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print(f"95% Confidence Interval for Population Mean Revenue: ${lower_bound:.2f} to ${upper_bound:.2f}")


# Based on the sample data, the 95% confidence interval for the population mean revenue is approximately $491.60 to $508.40. This means that we can be 95% confident that the true population mean revenue lies within this range.
```

```python
# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
# clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
# standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.




import numpy as np
from scipy.stats import ttest_1samp

# Given data
sample_mean = 8  # Sample mean decrease in blood pressure
population_mean = 10  # Hypothesized population mean decrease in blood pressure
sample_std_dev = 3  # Sample standard deviation of decrease in blood pressure
sample_size = 100  # Sample size

# Generate example data for the decrease in blood pressure
np.random.seed(42)
data = np.random.normal(loc=sample_mean, scale=sample_std_dev, size=sample_size)

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(data, population_mean)

# Define the significance level
significance_level = 0.05

# Check if the null hypothesis should be rejected or not
if p_value < significance_level:
    print("Null hypothesis rejected. The new drug does decrease blood pressure by 10 mmHg.")
else:
    print("Null hypothesis not rejected. There is insufficient evidence to claim that the new drug decreases blood pressure by 10 mmHg.")

```

```python
# Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
# standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
# is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
# pounds with a significance level of 0.01.




import numpy as np
from scipy.stats import ttest_1samp

# Given data
sample_mean = 4.8  # Sample mean weight of the products
population_mean = 5.0  # Population mean weight of the products
sample_std_dev = 0.5  # Sample standard deviation of the product weights
sample_size = 25  # Sample size

# Generate example data for the sample weights of the 25 products
np.random.seed(42)
data = np.random.normal(loc=sample_mean, scale=sample_std_dev, size=sample_size)

# Perform one-sample left-tailed t-test
t_statistic, p_value = ttest_1samp(data, population_mean)

# Define the significance level
significance_level = 0.01

# Check if the null hypothesis should be rejected or not
if p_value < significance_level and sample_mean < population_mean:
    print("Null hypothesis rejected. There is enough evidence to support that the true mean weight of the products is less than 5 pounds.")
else:
    print("Null hypothesis not rejected. There is insufficient evidence to claim that the true mean weight of the products is less than 5 pounds.")

```

```python
# Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
# 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
# score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
# groups are equal with a significance level of 0.01.





import numpy as np
from scipy.stats import ttest_ind

# Group 1 data
mean_group1 = 80
std_dev_group1 = 10
sample_size_group1 = 30

# Group 2 data
mean_group2 = 75
std_dev_group2 = 8
sample_size_group2 = 40

# Generate example data for the two groups using normal distributions
np.random.seed(42)
group1_scores = np.random.normal(loc=mean_group1, scale=std_dev_group1, size=sample_size_group1)
group2_scores = np.random.normal(loc=mean_group2, scale=std_dev_group2, size=sample_size_group2)

# Perform two-sample two-tailed t-test
t_statistic, p_value = ttest_ind(group1_scores, group2_scores)

# Define the significance level
significance_level = 0.01

# Check if the null hypothesis should be rejected or not
if p_value < significance_level:
    print("Null hypothesis rejected. There is enough evidence to support that the population means for the two groups are not equal.")
else:
    print("Null hypothesis not rejected. There is insufficient evidence to claim that the population means for the two groups are different.")
```
