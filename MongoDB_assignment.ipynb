````python
# Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
# MongoDB over SQL databases?



MongoDB is a popular document-oriented, NoSQL (non-relational) database system. It differs from traditional SQL databases by providing a flexible, schema-less data model and a distributed architecture designed to handle large-scale data storage and processing.

Non-relational databases, including MongoDB, are designed to store and manage unstructured or semi-structured data. Unlike SQL databases that use tables with fixed schemas and relations between them, non-relational databases use various data models such as key-value, document, columnar, or graph. They are suitable for scenarios where data structures may vary and evolve over time, and where horizontal scalability and high-performance requirements are crucial.

Advantages of using MongoDB (non-relational) over SQL databases:

1. **Schema Flexibility**: MongoDB allows dynamic and flexible schemas, where each document in a collection can have its own structure and fields. This makes it easier to handle evolving data requirements and supports agile development.

2. **Horizontal Scalability**: MongoDB's distributed architecture enables easy scaling horizontally by adding more servers or clusters to handle increased data and traffic loads. It provides built-in sharding capabilities to distribute data across multiple servers.

3. **High Performance**: MongoDB offers high-speed data access with its ability to store related data together in a single document. It also supports indexing, caching, and in-memory storage options, which can significantly improve query performance.

4. **Handling Big Data**: MongoDB is well-suited for handling large volumes of data, including big data scenarios. It provides automatic sharding and replication features, which ensure high availability, fault tolerance, and efficient data distribution.

5. **Frequent Writes**: If your application has a heavy write workload with frequent updates or inserts, MongoDB's flexible data model and native support for horizontal scaling can provide better performance and scalability compared to SQL databases.

6. **Real-time Analytics**: MongoDB's aggregation framework and support for MapReduce enable complex data processing and real-time analytics on large datasets, making it suitable for applications requiring advanced analytics capabilities.

7. **Geospatial Data**: MongoDB has excellent support for geospatial data and provides various geospatial indexing and querying features. It can efficiently handle location-based applications and geospatial data analysis.

It's important to note that the choice between MongoDB (non-relational) and SQL databases depends on the specific requirements and characteristics of your application. SQL databases excel in scenarios where structured data, complex relations, and ACID (Atomicity, Consistency, Isolation, Durability) transactions are critical, while MongoDB (non-relational) is preferred when flexibility, scalability, and high-performance handling of unstructured or semi-structured data are the priorities.
````

````python
# Q2. State and Explain the features of MongoDB.




MongoDB, as a leading NoSQL database system, offers several key features that make it popular among developers and organizations. Here are some of the prominent features of MongoDB:

1. **Flexible Document Model**: MongoDB uses a document-oriented data model where data is stored in flexible, JSON-like documents called BSON (Binary JSON). Documents can have varying structures, allowing for easy schema evolution and handling of diverse data types within a collection.

2. **Scalability and High Availability**: MongoDB is designed to scale horizontally, allowing you to distribute data across multiple servers or clusters. It supports automatic sharding, where data is partitioned and distributed across shards, enabling seamless scaling as data grows. Additionally, it provides built-in replication for high availability, ensuring data redundancy and failover.

3. **Indexing and Querying**: MongoDB supports a wide range of indexing options, including single field, compound, geospatial, and text indexes. This facilitates efficient querying and retrieval of data based on specific criteria. MongoDB's flexible query language allows for complex queries, aggregations, and full-text search capabilities.

4. **Ad Hoc Queries**: MongoDB allows ad hoc queries on dynamic schemas, enabling developers to explore and analyze data using a rich query language. Queries can be performed on specific fields, ranges, arrays, nested documents, and more, making it easy to retrieve and manipulate data.

5. **Replication and Fault Tolerance**: MongoDB provides replication capabilities through its replica set feature. It allows for automatic failover, data redundancy, and read scaling. Replica sets consist of primary and secondary nodes, where the primary handles all write operations, and the secondary nodes replicate the data and serve read operations.

6. **Aggregation Framework**: MongoDB's powerful aggregation framework enables data processing and analytics tasks. It supports various stages, including filtering, grouping, sorting, joining, and performing complex computations, allowing for advanced data analysis and real-time reporting.

7. **Geospatial and Text Search**: MongoDB has built-in support for geospatial indexing and querying, making it suitable for location-based applications and spatial data analysis. It also offers a robust text search feature that allows full-text indexing and search capabilities across collections.

8. **GridFS**: MongoDB's GridFS is a file storage system that enables efficient storage and retrieval of large files and binary data. It splits files into smaller chunks and stores them as separate documents, providing seamless integration of file storage with the database.

9. **Security**: MongoDB offers various security features, including authentication, authorization, role-based access control (RBAC), and transport layer encryption (TLS/SSL). It provides fine-grained access control at the collection and database levels, ensuring data security and compliance.

10. **Community and Ecosystem**: MongoDB has a large and active community, providing extensive resources, documentation, and support. It also offers a wide range of integrations, drivers, and tools for different programming languages and frameworks, making it easy to integrate MongoDB into existing tech stacks.

These features collectively make MongoDB a versatile and robust NoSQL database solution, capable of handling diverse data types, scaling to meet growing demands, and providing flexible querying and analytics capabilities.
````

````python
# Q4. Using the database and the collection created in question number 3, write a code to insert one record,
# and insert many records. Use the find() and find_one() methods to print the inserted record.


Certainly! Here's an example code that builds upon the previous code snippet to insert one record and multiple records into the MongoDB collection, and then uses the `find()` and `find_one()` methods to print the inserted records:


import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Access the database and collection
database = client["mydatabase"]
collection = database["mycollection"]

# Insert one record
document_one = {"name": "Alice", "age": 25}
inserted_one = collection.insert_one(document_one)
print("Inserted Record:")
print(collection.find_one({"_id": inserted_one.inserted_id}))

# Insert many records
documents_many = [
    {"name": "Bob", "age": 32},
    {"name": "Charlie", "age": 40},
    {"name": "Eve", "age": 28}
]
inserted_many = collection.insert_many(documents_many)
print("Inserted Records:")
for document in collection.find({"_id": {"$in": inserted_many.inserted_ids}}):
    print(document)

# Close the MongoDB connection
client.close()

In this code:

1. We establish a connection to MongoDB using `pymongo.MongoClient`.
2. We access the database and collection using the names you specified in the previous question.
3. We insert one record into the collection using `collection.insert_one()`. The inserted record is stored in the `inserted_one` variable.
4. We use `collection.find_one()` to retrieve and print the inserted record by passing its `_id` value obtained from `inserted_one.inserted_id`.
5. We insert multiple records into the collection using `collection.insert_many()` and store the inserted records in the `inserted_many` variable.
6. We use `collection.find()` to retrieve and print all the inserted records by querying for the `_id` values contained in `inserted_many.inserted_ids`.

Make sure to adapt the code to use the appropriate database and collection names you created in the previous question.

This code demonstrates the insertion of one record and multiple records into a MongoDB collection and showcases the usage of `find()` and `find_one()` methods to retrieve the inserted records.
````

````python
# Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
# demonstrate this.



In MongoDB, the `find()` method is used to query the database and retrieve documents that match certain criteria. It allows you to specify filters, projections, sorting, and other options to customize the query results. The `find()` method returns a cursor object that can be iterated to access the matching documents.

Here's a simple code example that demonstrates how to use the `find()` method to query a MongoDB database:


import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Access the database and collection
database = client["mydatabase"]
collection = database["mycollection"]

# Find all documents in the collection
cursor = collection.find()

# Iterate and print the documents
print("All Documents:")
for document in cursor:
    print(document)

# Find documents that match a specific criteria
query = {"age": {"$gt": 30}}  # Find documents where "age" field is greater than 30
cursor = collection.find(query)

# Iterate and print the matching documents
print("Matching Documents:")
for document in cursor:
    print(document)

# Close the MongoDB connection
client.close()

In this code:

1. We establish a connection to MongoDB using `pymongo.MongoClient`.
2. We access the database and collection using the appropriate names.
3. We use `collection.find()` without any arguments to retrieve all documents in the collection. This returns a cursor object.
4. We iterate over the cursor and print each document, displaying all the documents present in the collection.
5. We define a query using a filter criteria to find documents where the `"age"` field is greater than 30. In this case, we use `{"age": {"$gt": 30}}` to specify the criteria.
6. We use `collection.find(query)` with the defined query to retrieve documents that match the specified criteria. This returns a cursor object.
7. We iterate over the cursor and print each matching document, displaying only the documents that meet the filter criteria.

Note that the query in the example uses the `$gt` operator to find documents with an "age" field greater than 30. MongoDB provides a wide range of query operators and options to perform complex queries based on various criteria.

Make sure to adjust the code according to your database and collection names.
````

````python
# Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.



In MongoDB, the `sort()` method is used to specify the sorting order for the query results. It allows you to sort documents based on one or more fields in ascending or descending order. The `sort()` method is typically used in conjunction with the `find()` method to retrieve sorted query results.

The `sort()` method takes a document as its argument, where each field represents the sorting order. The field's value can be `1` for ascending order or `-1` for descending order. Multiple fields can be included in the sort document to specify a compound sort order.

Here's an example that demonstrates the usage of the `sort()` method in MongoDB:


import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Access the database and collection
database = client["mydatabase"]
collection = database["mycollection"]

# Find and sort documents in ascending order based on the "age" field
cursor = collection.find().sort("age", 1)

# Print the sorted documents
print("Sorted Documents (Ascending):")
for document in cursor:
    print(document)

# Find and sort documents in descending order based on the "name" field
cursor = collection.find().sort("name", -1)

# Print the sorted documents
print("Sorted Documents (Descending):")
for document in cursor:
    print(document)

# Close the MongoDB connection
client.close()


In this code:

1. We establish a connection to MongoDB using `pymongo.MongoClient`.
2. We access the database and collection using the appropriate names.
3. We use `collection.find()` to retrieve all documents in the collection.
4. We use the `sort()` method with the field name `"age"` and the value `1` to sort the documents in ascending order based on the "age" field.
5. We iterate over the cursor and print each document, displaying the sorted documents in ascending order of the "age" field.
6. We use the `sort()` method again, but this time with the field name `"name"` and the value `-1` to sort the documents in descending order based on the "name" field.
7. We iterate over the cursor and print each document, displaying the sorted documents in descending order of the "name" field.

You can modify the code and specify multiple fields and their respective sorting orders in the `sort()` method to perform compound sorting based on multiple criteria.

Ensure that you adapt the code according to your database and collection names.
````

````python
# Q7. Explain why delete_one(), delete_many(), and drop() is used.




In MongoDB, the `delete_one()`, `delete_many()`, and `drop()` methods are used to remove documents and collections from a database. Here's an explanation of when and how each method is used:

1. **delete_one(filter)**: The `delete_one()` method is used to delete a single document that matches a specific filter. It takes a filter as an argument, which defines the criteria for identifying the document to be deleted. If multiple documents match the filter, only the first matching document encountered will be deleted.

   Example usage:
   ```python
   # Delete a document with a specific name
   collection.delete_one({"name": "Alice"})
   ```

2. **delete_many(filter)**: The `delete_many()` method is used to delete multiple documents that match a specific filter. It takes a filter as an argument, similar to `delete_one()`, but it deletes all documents that match the given filter.

   Example usage:
   
   # Delete all documents with age greater than 30
   collection.delete_many({"age": {"$gt": 30}})
   

3. **drop()**: The `drop()` method is used to remove an entire collection from a database. It permanently deletes the collection, including all its documents and indexes. This method is typically used when you want to completely remove a collection and its associated data from the database.

   Example usage:

   # Drop the entire collection
   collection.drop()


Each of these methods serves a different purpose:

- `delete_one()` is used when you want to remove a specific document that matches a given filter. It is suitable when you need to delete a single record or the first occurrence of a specific document.

- `delete_many()` is used when you want to delete multiple documents that match a given filter. It is useful when you need to remove multiple records that satisfy a certain condition.

- `drop()` is used when you want to remove an entire collection from the database. It is helpful when you need to completely delete a collection, including all its documents and indexes.

It is important to exercise caution when using these methods, especially `delete_many()` and `drop()`, as they can permanently delete data from the database. Always double-check the filters and ensure that you have the necessary backups or data protection measures in place before performing such operations.
````
